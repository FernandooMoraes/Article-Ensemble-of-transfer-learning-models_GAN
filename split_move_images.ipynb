```python
rotulos_train=rotulos.iloc[:810]
rotulos_test=rotulos.iloc[810:900]
```

```python
rotulos_train["1"].iloc[0]
```

```python
diretorio_dest = '/content/drive/MyDrive/kfold/fold'
diretorio_orig = '/content/imagensprocessadas/'
```

```python
def mover_imagens(diretorio_dest,diretorio_orig):
  for j in range (9):
    rot = rotulos_train.columns[j]
    for i in range (810):
        dados = rotulos_train[rot].iloc[i]
        imagem_o = diretorio_orig + dados + ".jpg"
        imagem_s = diretorio_dest + rot + "/train/" + dados + ".jpg"

        shutil.copy(imagem_o,imagem_s)
```

```python
mover_imagens(diretorio_dest,diretorio_orig)
```

```python
def mover_imagens(diretorio_dest,diretorio_orig):
  for j in range (9):
    rot = rotulos_test.columns[j]
    for i in range (90):
        dados = rotulos_test[rot].iloc[i]
        imagem_o = diretorio_orig + dados + ".jpg"
        imagem_s = diretorio_dest + rot + "/test/"+ dados + ".jpg"

        shutil.copy(imagem_o,imagem_s)
```

```python
mover_imagens(diretorio_dest,diretorio_orig)
```
